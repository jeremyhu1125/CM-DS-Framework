```python
import math
def BAPs(x,Ex,En):
    tmp = (x-Ex)**2/2/En**2
    return np.exp(-tmp)
```

```python
BAPs_df= pd.DataFrame(columns= pd.MultiIndex.from_product([['A', 'B', 'C','D'],cols] ),index =merge_df.indicator)
```

```python
BAPs_df['A']=np.array(merge_df[cols].apply(BAPs,Ex=merge_df.A_Ex,En=merge_df.A_En))
```

```python
BAPs_df['B']=np.array(merge_df[cols].apply(BAPs,Ex=merge_df.B_Ex,En=merge_df.B_En))
```

```python
BAPs_df['C']=np.array(merge_df[cols].apply(BAPs,Ex=merge_df.C_Ex,En=merge_df.C_En))
```

```python
BAPs_df['D']=np.array(merge_df[cols].apply(BAPs,Ex=merge_df.D_Ex,En=merge_df.D_En))
```

```python
BAPS = BAPs_df.T
BAPS.index.names = ['Level', 'Item']
```

```python
BAPS.head()
```

```python
BAPS = BAPS.swaplevel('Level', 'Item').T
```

```python
BAP_fin = pd.DataFrame()
for i,col in enumerate(cols):
    tmp = BAPS[col]
    col_tmp = ['A'+str(i+1),'B'+str(i+1),'C'+str(i+1),'D'+str(i+1)]   
    tmp.columns = col_tmp
    theta ='mtheta'+str(i+1)
    tmp[theta] = 1-tmp.sum(axis=1)
    BAP_fin  = pd.concat([BAP_fin,tmp],axis =1)
```

```python
def construct_mass(df,indicator,n):
    mass_dict = dict()
    level = ['a','b','c','d','abcd']
    tmpdf = df.loc[indicator]
    for i in range(0,n):
        key = indicator +str(i)
        masstmp= MassFunction()
        for j,d in enumerate(level):
            p = 5*i+j
            masstmp[d] =tmpdf[p]
        mass_dict[key]  =  masstmp
        if i==0:
            res=masstmp
        else:
            res = res&masstmp
    return mass_dict,res
```

```python
Q_ACC_mass,Q_ACC_res= construct_mass(condf,'Q_ACC',3)
```
